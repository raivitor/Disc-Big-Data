```python
# Import pandas as pd
import pandas as pd

# Import the fertility.csv data: data
data = pd.read_csv("bbc.csv", encoding = 'latin2')

df_filter = []
for index, row in data.iterrows():
    try:
        if (int(row["Recognition Score"])<=100) and (int(row["Temporal Memory Score"])<=100):
            df_filter.append(row)
    except:
        continue

df_filter = pd.DataFrame(df_filter)        

len(df_filter)
df_filter.to_csv("bbcversao2.csv",  encoding = 'latin2')
```

```python
# Import pandas as pd
import pandas as pd
# Import Scatter, output_notebook, and show from bokeh.charts
from bokeh.charts import Scatter, output_notebook, show

data = pd.read_csv("bbcversao2.csv", encoding = 'latin2')

data = data[data['sono'].notnull()]
data = data[data['Temporal Memory Score'].notnull()]

# Make a scatter plot: p
p = Scatter(data, x='sono', y='Temporal Memory Score', title='', plot_width= 900)

# Set the x-axis label
p.xaxis.axis_label = 'sono'

# Set the y-axis label
p.yaxis.axis_label = 'Temporal Memory Score'

# Call the output_notebook() 
output_notebook()
show(p)
```
