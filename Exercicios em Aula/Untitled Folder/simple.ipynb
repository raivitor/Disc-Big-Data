```python
from bs4 import BeautifulSoup
import requests

url ="http://dataquestio.github.io/web-scraping-pages/simple.html"

response = requests.get(url)

content = response.content

soup = BeautifulSoup(content, "html.parser")
```

```python
from bs4 import BeautifulSoup
import requests

url ="http://globo.com"
response = requests.get(url)
content = response.content
soup = BeautifulSoup(content, "html.parser")

a_tags = soup.find_all('a')

for link in a_tags:
    print(link.get('href'))
```

```python
from bs4 import BeautifulSoup
import requests

url ="http://dataquestio.github.io/web-scraping-pages/simple_ids.html"
response = requests.get(url)
content = response.content
soup = BeautifulSoup(content, "html.parser")

fp = soup.find_all(id="first")[0]
print(fp)
```
